```python
#get eigen values and sample labels
pca_df=pd.read_csv("pca.eigenvec", sep=" ", header=None)
pca_df.columns=["sample_name", "x"] + list(range(1, 21))
pca_df=pca_df.drop("x", axis=1)

# #now add sample info (including location data)
info_df=pd.read_csv("{}/data/seq_and_sample_docs/all_sh_sb_sample_data.csv".format(proj_dir), sep=",") 

#merge the dataframes
pca_df=pca_df.merge(info_df, how='left', on='sample_name')

#fix the country which contains some float NaNs
pca_df["country"] = pca_df["country"].astype(str)

#set marker colors based on the country
cmap = cm.get_cmap('nipy_spectral')
colors={}

countries = sorted(pca_df["country"].unique().astype(str))

i=0
for country in countries:
    colors[country]=cmap(np.linspace(0,1,len(countries))[i])
    i+=1
```

```python
#code modified from https://predictivehacks.com/k-means-elbow-method-code-for-python/

#get pc1,pc2
pc1_pc2=pca_df.iloc[:,1:3].values.tolist()

distortions = []
K = range(1,10)

for k in K:
    kmeanModel = KMeans(n_clusters=k, n_init='auto')
    kmeanModel.fit(pc1_pc2)
    distortions.append(kmeanModel.inertia_)
    
    
plt.figure()
plt.plot(K, distortions, 'bx-')
plt.xlabel('k')
plt.ylabel('Distortion')
plt.title('The Elbow Method showing the optimal k')

plt.savefig("kmeans_elbow.png", dpi=300)
plt.savefig("kmeans_elbow.svg")

plt.show()
```

```python
fig, ax = plt.subplots()

#visualize the three groups
kmeans=KMeans(n_clusters=3, random_state=0, n_init='auto').fit_predict(pc1_pc2)
pca_df["kmeans_group"]=kmeans

k_colors={0:"red", 1:"green", 2:"blue" }

for index, row in pca_df.iterrows():
    x=row[1]
    y=row[2]
    color=k_colors[kmeans[index]]
    plt.scatter(x, y, color=color, edgecolor="black", linewidths=0.3, s=60)
    

#add rectangles
for k in [0,1,2]:
    x1=pca_df[pca_df["kmeans_group"]==k][1].min()
    x2=pca_df[pca_df["kmeans_group"]==k][1].max()

    y1=pca_df[pca_df["kmeans_group"]==k][2].min()
    y2=pca_df[pca_df["kmeans_group"]==k][2].max()

    rect = mpatches.Rectangle((x1-.01, y1-.01), (x2-x1)+.02, (y2-y1)+.02, linewidth=1, edgecolor='grey', facecolor='none')
    
    # Add the patch to the Axes
    ax.add_patch(rect)

plt.show()

```

```python
fst_dfs={}

for pop1, pop2 in [("sh_nw", "sh_se"), 
                   ("sh_nw", "sb"), 
                   ("sh_se", "sb"), 
                   ("sh", "sb")]:
    
    comp = f"{pop1}_v_{pop2}"
    
    chrs=[]
    fsts=[]
    starts=[]
    stops=[]

    tmp_df = pd.DataFrame()

    for chrom in uniq_chroms:
        mask = (chroms == chrom)

        chrom_gt  = gt[mask]
        chrom_pos = poss[mask]

        #get windows:
        indices = np.arange(0, len(chrom_pos), 500)  # We use the length of chrom_pos to generate indices

        # Initialize empty lists to hold the start and stop positions
        chr_starts = []
        chr_stops = []

        # Create start and stop positions using the indices
        for i in indices[:-1]: # Exclude the last index to avoid short windows
            chr_start = chrom_pos[i]  # Start position is the current index
            chr_starts.append(chr_start)

            # Stop position is the index 500 steps ahead, it always exists because we excluded the last index
            chr_stop = chrom_pos[i + 500]
            chr_stops.append(chr_stop)

        # Convert lists to numpy arrays for consistency with previous code
        chr_starts = np.array(chr_starts)
        chr_stops = np.array(chr_stops)

        fst = allel.moving_weir_cockerham_fst(chrom_gt, [kmeans_pop_idxs[pop1], kmeans_pop_idxs[pop2]], size=500)

        chrs.extend([chrom] * len(fst))
        fsts.extend(fst)
        starts.extend(chr_starts)
        stops.extend(chr_stops)

    tmp_df["chr"] = chrs
    tmp_df["start"]=starts
    tmp_df["stop"]=stops
    tmp_df[comp]=fsts

    tmp_df.to_csv(f"{comp}_genome_wide_fst.csv", header=True, sep=",", index=False)
    
    fst_dfs[comp]=tmp_df
```

```python
def read_samples(filename):
    return np.genfromtxt(filename, delimiter=',', dtype=str)

def get_sample_indices(samples, ref_samples):
    return [samples.index(ref_sample) for ref_sample in ref_samples]

def read_vcf(vcf_file):
    vcf = VCF(vcf_file)
    snp_ids = []
    chroms = []
    positions = []
    genotype_data = []

    for record in vcf:
        snp_ids.append(record.ID)
        chroms.append(record.CHROM)
        positions.append(record.POS)
        genotype_data.append(record.genotypes)
    
    return snp_ids, chroms, positions, genotype_data

def convert_genotypes(genotype_data):
    converted_genotypes = [
        [f"{genotype[0]}{genotype[1]}" if genotype[0] != -1 and genotype[1] != -1 else ".." for genotype in row]
        for row in genotype_data
    ]
    return list(map(list, zip(*converted_genotypes)))

def extract_genotypes(samples, genotypes, ref_samples):
    gts = {sample: ht_1 for sample, ht_1 in zip(samples, genotypes)}
    return {key: gts[key] for key in ref_samples if key in gts}

# def select_haplotype(pop_gts):
#     breeder = random.choice(list(pop_gts.keys()))
#     hap = random.choice([0, 1])
#     return [allele[hap] for allele in pop_gts[breeder]]

def select_haplotype(pop_gts):
    
    indiv = random.sample(list(pop_gts.keys()), k=1)[0]
    gt=pop_gts[indiv]
    chrom=random.choice([0, 1])
    ht=[allele[chrom] for allele in gt]
    sample_id=f"{indiv}_h{chrom}"
    
    return sample_id, ht

def mate(pop_a_gts, pop_b_gts):
    pop_a_indv = ""
    pop_b_indv = ""
    
    while pop_a_indv == pop_b_indv:
        pop_a_id, pop_a_seq = select_haplotype(pop_a_gts)
        pop_b_id, pop_b_seq = select_haplotype(pop_b_gts)

        pop_a_indv = pop_a_id.split("_h")[0] 
        pop_a_indv = pop_a_id.split("_h")[1]

    offspring_gt = [str(a) + str(b) for a, b in zip(pop_a_seq, pop_b_seq)]
    offspring_name=f"{pop_a_id}-{pop_b_id}"
    
    return offspring_name, offspring_gt

def recombination_simulation(chromosome_length_bp, recombination_rate_cM):
    # Convert recombination rate from cM to probability per bp
    recombination_rate_bp = 0.01/recombination_rate_cM

    # Initialize an empty list to store recombination positions
    recombination_positions = []

    # Iterate along the chromosome
    for position in range(chromosome_length_bp):
        
        # Calculate the probability of recombination at this position
        recombination_probability = recombination_rate_bp

        # Generate a random number between 0 and 1
        random_number = np.random.random()

        # If the random number is smaller than the recombination probability, a recombination event occurred
        if random_number < recombination_probability:
            recombination_positions.append(int(position))

    return recombination_positions

def identify_breakpoints(chromosome_lengths, recombination_rate_cM):
    breakpoints={}
    for chrom in chrom_lengths.keys():
        breakpoints[chrom] = recombination_simulation(chrom_lengths[chrom], recombination_rate_cM)
        
    return breakpoints

def group_genotypes_by_chromosome(offspring_genotype, chroms, positions):
    grouped_genotypes = defaultdict(list)
    for genotype, chrom, pos in zip(offspring_genotype, chroms, positions):
        grouped_genotypes[chrom].append((genotype, pos))
    return grouped_genotypes

def recombine_genotypes(offspring_genotype, chroms, positions, breakpoints):
    grouped_genotypes = group_genotypes_by_chromosome(offspring_genotype, chroms, positions)
    
    for chrom, position in breakpoints.items():
        for position in positions:
            mask = ((df['chrom'] == chrom) & (df['pos'] >= position))
            df.loc[mask, ['h1', 'h2']] = df.loc[mask, ['h2', 'h1']].values

        df["gt"] = df.apply(lambda x: x['h1'] + x['h2'], axis=1)

    return list(df["gt"])
    
def produce_offspring(pop_a_gts, pop_b_gts, chrom_lengths, chroms, positions, recomTF, recomb_cM):
    break_points=[]
    indv, gt = mate(pop_a_gts, pop_b_gts)
    
    if recomTF==True:
        break_points = identify_breakpoints(chrom_lengths, recomb_cM)
        gt = recombine_genotypes(gt, chroms, positions, break_points)
    
    return indv, gt, break_points

def prog_bar(iterations):
    progress = i / (int(iterations) - 1)
    bar_length = 50
    filled_length = int(progress * bar_length)
    bar = "#" * filled_length + "-" * (bar_length - filled_length)
    print(f"\rProgress: [{bar}] {progress*100:.1f}%", end="")
```
